# An introduction to Data Visualisation

## With Python

## Good Data Visualisation

- ### Objective and unbiased representation of the data

- ### Should deliver meaningful results and illustrate something useful about the data

- ### All information required to understand it should be available in the legend and/or caption (but may require domain specific knowledge)

- ### Can also be used to explore your data - figures and plots for you and your collaborators, not for publication!

## What's needed for **Good Data Visualisation**?

- ### Evidence-based best practises for objective data representation

- ### Technical skills in tools to build graphics

- ### Domain-specific knowlege of underlying dataset

- ### Creativity and subjective taste

## Who is your intended audience?



- ### General public, non-research audiences
    - Sci-art
    - Infographics

- ### Non-domain specific, but technical audiences

- ### Domain specific research audiences

- ### Your collaborators

- ### You!

# Plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dots = sns.load_dataset("dots")
dots = dots.rename(columns={"align":"Align", "choice":"Choice", "time":"Time", "coherence": "Coherence", "firing_rate": "Firing Rate"})

dots
```

```python

# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="Time", y="Firing Rate",
    hue="Coherence", size="Choice", col="Align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)

plt.show()
```

```python
with sns.plotting_context("talk"):
    # Plot the lines on two facets
    sns.relplot(
        data=dots,
        x="Time", y="Firing Rate",
        hue="Coherence", size="Choice", col="Align",
        kind="line", size_order=["T1", "T2"], palette=palette,
        height=5, aspect=.75, facet_kws=dict(sharex=False),
    )

    plt.show()

print("With context set to 'talk'")
```

```python
with sns.plotting_context("paper"):
    # Plot the lines on two facets
    sns.relplot(
        data=dots,
        x="Time", y="Firing Rate",
        hue="Coherence", size="Choice", col="Align",
        kind="line", size_order=["T1", "T2"], palette=palette,
        height=5, aspect=.75, facet_kws=dict(sharex=False),
    )

    plt.show()

print("With context set to 'paper'")
```
